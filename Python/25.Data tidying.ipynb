#### 8. Replacing (substituting) missing data with other data

* 1 Replacing wiht the mean value 


```python
import pandas as pd 
import seaborn as sns

df = sns.load_dataset('titanic', cache = True)

df['age'].head(10) # When checking the data in the 'age' column, NaN value are found.
```

```python
avg_age = df['age'].mean(axis = 0)  # Calculate the mean value of the age data and store it in 'avg_age'. 
                                    # It represents the average value of the 'age' column data.
avg_age
```

```python
median_age = df['age'].median(axis = 0) # The medin value can be calculate using the 'median()'method.
median_age
```

* Using the 'fillna()' method, we can replace NaN data elements  with the mean values. Let's use the previously calculated
  'avg _ age' to replace them.

```python
df['age'].fillna(avg_age, inplace = True)
```

```python
df['age'].head(10) # We can see that missing data has been replaced with the mean value.
```

```python
df.isnull().sum()
```

* 2) Replacing with the most freqient value
     * Let's search for the most frequently boarded city name among passengers and replace the missing data in the 'embark_town' column with that data.

```python
import pandas as pd 
import seaborn as sns

df = sns.load_dataset('titanic', cache = True)
```

```python
town_count = df['embark_town'].value_counts(dropna = True) # vlaue_count return a series that shows the count of unique occureences of each row in the 
#column .It excludes missing data.

town_count
```

```python
type(town_count)
```

```python
most = town_count.idxmax() # idxam() returns the index where the maximun value first occurs in the requested axis.
most
```

```python
df['embark_town'].fillna(most, inplace = True) # fillna() method we replace NaN data elemets with the most frequent city 
 
df.isnull().sum()
```
